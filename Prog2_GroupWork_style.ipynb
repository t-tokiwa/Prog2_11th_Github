Turtleライブラリを使って以下のような図を作成してください。

長さは毎回長さが0.9倍されるようにする。
角度は30度ずつ内側へ渦を巻くように。

```python
!pip3 install ColabTurtlePlus
```

```python
# ここに解答または解答例を記入する
from ColabTurtlePlus.Turtle import *
clearscreen()
setup(300,300)
showborder()

turtle1=Turtle()
turtle1.shape('turtle')
turtle1.speed(10)
turtle1.pencolor('red')

turtle1.penup()
turtle1.goto(145,0)
turtle1.setheading(90)
turtle1.pendown()
n=30
length=60

for i in range(n):
  turtle1.forward(length)
  turtle1.left(30)
  length*=0.9

turtle1.penup()
turtle1.goto(0,0)

turtle1.pencolor('blue')
turtle1.penup()
turtle1.goto(-145,0)
turtle1.setheading(-90)
turtle1.pendown()
n=30
length=60

for i in range(n):
  turtle1.forward(length)
  turtle1.left(30)
  length*=0.9

turtle1.penup()
turtle1.goto(0,0)
```

## 問題3解説
解説はここに記入する．
