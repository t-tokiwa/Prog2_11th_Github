# 問題1（初級）

ここに問題を記入する
x を−1.0以上1.0以下で，0.01刻みの配列として定義し， $y=2x^2+3$ を図示せよ
・田村亘輝

```python
# ここに解答または解答例を記入する
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-1.0, 1.01, 0.01) # -1.0 <= x < 1.01
y = 2*x**2 + 3 # **はベキ乗．

plt.plot(x,y)
plt.show()
```

# 問題2（中級）

次の関数を１つの図に図示してください．ただし，変数 x は −10<=x<=10 の範囲で等間隔の1,000個とします．

y=5x+4

y=3x^2+x−50

* 大下　冬夢

```python
# ここに解答または解答例を記入する
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000) # -10 <= x <= 10, 1000点
y1 = 5 * x +4
y2 = 3 * x ** 2 + x -50


plt.plot(x,y1)
plt.plot(x,y2)

plt.grid()
plt.show()
```

# 問題3（上級）

ここに問題を記入する．
素因数分解するプログラムをつくれ！
市岡 優臣
* 問題提案者の氏名を問題末尾に記入して下さい．複数人で作成した場合は連名で構いません．

```python
# ここに解答または解答例を記入する

a = int(input('数字を入力'))
b = a
num = []
counter = []
for i in range(2,a+1):
  count = 0
  if a % i == 0:
    while True:
      if a % i == 0:
        a /= i
        count += 1
      else:
        break
    num.append(i)
    counter.append(count)
text = str(b) + 'を素因数分解すると'
for n in num:
  text += str(n) + 'が' + str(counter[num.index(n)]) + '個、'
print(text)

```
